# Python function to generate an automated EDA report using Pandas Profiling.

```python
# Cell: ydata-profiling example
import seaborn as sns
from IPython.display import IFrame
from ydata_profiling import ProfileReport

df = sns.load_dataset("titanic")

def eda_ydata_profiling(df, output_path="ydata_profiling_titanic.html", **profile_kwargs):
    profile = ProfileReport(df, **profile_kwargs)
    profile.to_file(output_path)
    return output_path

# Example: generate & show report
out = eda_ydata_profiling(df, output_path="ydata_profiling_titanic.html", title="YData Profiling - Titanic", explorative=True)
IFrame(out, width="100%", height=700)
```
